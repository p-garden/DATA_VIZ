## 2. 1인가구의 연령대별 비중

```python
years = df.iloc[0, 2:].reset_index(drop=True).apply(lambda x: str(x).split('.')[0]).astype(int)  # Extract unique year values
age_groups = df.iloc[2, 2:].reset_index(drop=True)  # Third row contains age group information
data = df.iloc[3, 2:].reset_index(drop=True)  # Fourth row contains household counts

# Create a new DataFrame for the cleaned data
cleaned_df = pd.DataFrame({
    '연도': years,
    '연령대': age_groups,
    '가구수': data
})

# Drop rows with NaN values
cleaned_df = cleaned_df.dropna()
cleaned_df = cleaned_df[cleaned_df['연령대'] != '소계']
cleaned_df = cleaned_df[cleaned_df['연도'] >= 2015]

# Convert '연도' to int and '가구수' to numeric
cleaned_df['연도'] = cleaned_df['연도'].astype(int)
cleaned_df['가구수'] = pd.to_numeric(cleaned_df['가구수'])

# Display the cleaned dataframe
print(cleaned_df)
```

```python
pivot_df = cleaned_df.pivot_table(values='가구수', index='연도', columns='연령대', aggfunc='sum')

# Plotting the line chart for each age group over the years
plt.figure(figsize=(14, 8))
for age_group in pivot_df.columns:
    plt.plot(pivot_df.index, pivot_df[age_group], label=age_group)

plt.title('연도별 각 연령대별 1인가구 가구수 추이')
plt.xlabel('연도')
plt.ylabel('가구수')
plt.legend(title='연령대', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 3. 1인가구의 취업비중

```python
file_path = 'data/1인_취업가구_현황_20241108190218.csv'
df = pd.read_csv(file_path)

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager

years = list(range(2015, 2024))  # Years from 2015 to 2023
employment_proportion = df.iloc[2, 4:4 + len(years) * 2:2].reset_index(drop=True)  # Extract employment proportion data (even columns)

# Add the value 60.4 as the first element of employment_proportion
employment_proportion = pd.concat([pd.Series([60.4]), employment_proportion], ignore_index=True)

# Convert employment proportion to numeric values
employment_proportion = pd.to_numeric(employment_proportion)

# Set the font to 'Malgun Gothic' using the specified path for proper Korean character rendering
font_path = "C:/Windows/Fonts/malgun.ttf"  # Windows에서 사용하는 한글 폰트 경로
font = font_manager.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font

# Plotting the employment proportion over the years
plt.figure(figsize=(10, 6))
plt.plot(years, employment_proportion, marker='o', linestyle='-', color='blue', label='취업 가구 비중 (%)')
plt.title('2015~2023년 1인 취업 가구 비중 추이')
plt.xlabel('연도')
plt.ylabel('취업 가구 비중 (%)')
plt.xticks(years)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

```
