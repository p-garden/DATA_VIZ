# 1. 1인가구 관련 데이터 시각화

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_excel('data/1인가구(연령별)_20241107131601.xlsx', header=None)
```

## 1. 1인가구수 추이

```python
# 1. 첫 번째 행을 열 이름으로 설정
df.columns = df.iloc[0]
df = df.drop(df.index[0])
# 2. 데이터 변형 (melt)
df_melted = df.melt(id_vars=['자치구별(1)', '성별(1)'], var_name='연도', value_name='가구수')

# 3. 연도와 나이대 구분
def assign_age_and_year(col):
    """열 이름으로 연도와 나이대를 분리."""
    if '.' in col:
        year, age = col.split('.', 1)
        return int(year), age
    else:
        return int(col), '소계'

df_melted[['연도', '나이대']] = df_melted['연도'].apply(assign_age_and_year).apply(pd.Series)

# 4. 정리된 데이터프레임 생성
df_final = df_melted[['연도', '나이대', '성별(1)', '가구수']]

# 5. 열 이름 변경
df_final.columns = ['연도', '나이대', '성별', '가구수']

# '가구수' 열에서 숫자가 아닌 데이터를 제외하고, 연도별 1인가구 가구수의 합계를 계산 (2016~2023년 선택)
df_filtered = df_final[(df_final['성별'] == '계') & (df_final['연도'] >= 2016) & (df_final['연도'] <= 2023)]
df_filtered['가구수'] = pd.to_numeric(df_filtered['가구수'], errors='coerce')
one_person_households = df_filtered[df_filtered['성별'] == '계'].groupby('연도')['가구수'].sum()

one_person_households = one_person_households/2
# 결과 출력
print(one_person_households)
```

```python
# 시각화
plt.figure(figsize=(10, 6))
plt.plot(one_person_households.index, one_person_households.values, marker='o', color='b', linestyle='-', label='1인가구 가구수')
plt.title('2016~2023년 연도별 1인가구수 변화 추이')
plt.xlabel('연도')
plt.ylabel('가구수')
plt.xticks(rotation=45)
plt.grid(True)

# y축에 쉼표 추가
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.legend()
plt.tight_layout()

# 그래프 출력
plt.show()
```

## 2. 1인가구의 연령대별 비중

```python
df = pd.read_excel('data/1인가구(연령별)_20241107131601.xlsx', header=None)

years = df.iloc[0, 2:].reset_index(drop=True).apply(lambda x: str(x).split('.')[0]).astype(int)  # Extract unique year values
age_groups = df.iloc[2, 2:].reset_index(drop=True)  # Third row contains age group information
data = df.iloc[3, 2:].reset_index(drop=True)  # Fourth row contains household counts

# Create a new DataFrame for the cleaned data
cleaned_df = pd.DataFrame({
    '연도': years,
    '연령대': age_groups,
    '가구수': data
})

# Drop rows with NaN values
cleaned_df = cleaned_df.dropna()
cleaned_df = cleaned_df[cleaned_df['연령대'] != '소계']
cleaned_df = cleaned_df[cleaned_df['연도'] >= 2015]

# Convert '연도' to int and '가구수' to numeric
cleaned_df['연도'] = cleaned_df['연도'].astype(int)
cleaned_df['가구수'] = pd.to_numeric(cleaned_df['가구수'])

# Display the cleaned dataframe
print(cleaned_df)
```

```python
pivot_df = cleaned_df.pivot_table(values='가구수', index='연도', columns='연령대', aggfunc='sum')

# 데이터 재정립: 10살 단위로 연령대 그룹화
age_groups = {
    '20대': ['20~24세', '25~29세'],
    '30대': ['30~34세', '35~39세'],
    '40대': ['40~44세', '45~49세'],
    '50대': ['50~54세', '55~59세'],
    '60대': ['60~64세', '65~69세'],
    '70대': ['70~74세', '75~79세'],
    '80대 이상': ['80~84세', '85세이상']
}

# 새로운 데이터프레임 생성: 10살 단위로 그룹화된 데이터
pivot_df_grouped = pd.DataFrame()

for group, ages in age_groups.items():
    pivot_df_grouped[group] = pivot_df[ages].sum(axis=1)

# Plotting the line chart for each age group over the years
plt.figure(figsize=(14, 8))

# Define markers and colors for better visibility
markers = ['o', 's', 'D', '^', 'v', '>', '<']
colors = plt.cm.viridis(np.linspace(0, 1, len(pivot_df_grouped.columns)))

# Iterate over each age group to plot
for i, (age_group, color) in enumerate(zip(pivot_df_grouped.columns, colors)):
    plt.plot(
        pivot_df_grouped.index, 
        pivot_df_grouped[age_group], 
        label=age_group,
        color=color,
        marker=markers[i % len(markers)],  # Cycle through markers
        linewidth=2
    )

plt.title('연도별 각 연령대별 1인가구 가구수 추이')
plt.xlabel('연도')
plt.ylabel('가구수')
plt.xticks(rotation=45)
plt.grid(True)

# y축에 쉼표 추가
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.legend(title='연령대', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
file_path_3 = 'data/월평균_간편식_HMR__구입_지출액_20241109174813.csv'
df_hmr = pd.read_csv(file_path_3)

# Extract relevant data for visualization
years_hmr = ['2018', '2019', '2020', '2021', '2022', '2023']

# Extract average spending for each year from row labeled '소계'
spending_hmr = df_hmr.loc[df_hmr['특성별(2)'] == '소계'].iloc[:, 2:].values.flatten().astype(float)

# Plotting the line chart for average spending over the years
plt.figure(figsize=(10, 6))
plt.plot(years_hmr, spending_hmr, marker='o', linestyle='-', color='green', label='전체 평균 지출액')
plt.title('연도별 전체 월평균 간편식 구입 지출액 추이')
plt.xlabel('연도')
plt.ylabel('월평균 지출액 (원)')
plt.xticks(years_hmr, rotation=45)
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# Extract average spending for each household size
# Extract average spending for each household size
```

```python
df_hmr = df_hmr.iloc[1:, :]

# Rename the columns to fix header alignment
df_hmr.columns = ['특성별(1)', '특성별(2)', '2018', '2019', '2020', '2021', '2022', '2023']
# Remove '소계' row
df_hmr = df_hmr[df_hmr['특성별(2)'] != '소계']

# Replace '-' with NaN and ensure numeric conversion
df_hmr.replace('-', np.nan, inplace=True)
df_hmr.iloc[:, 2:] = df_hmr.iloc[:, 2:].astype(float)

# Combine '4인' and '5인 이상' into '4인 이상' for 2018-2020
df_4_more_2018_2020 = df_hmr[(df_hmr['특성별(2)'] == '4인') | (df_hmr['특성별(2)'] == '5인 이상')]
df_4_more_avg_2018_2020 = df_4_more_2018_2020[years_hmr[:3]].replace('-', np.nan).astype(float).mean().to_frame().T
df_4_more_avg_2018_2020['특성별(2)'] = '4인 이상'
df_4_more_avg_2018_2020.columns = ['2018', '2019', '2020', '특성별(2)']  # Ensure proper column alignment

# Extract '4인 이상' for 2021-2023
df_4_more_2021_2023 = df_hmr[df_hmr['특성별(2)'] == '4인 이상'][years_hmr[3:]].replace('-', np.nan).astype(float).copy()
df_4_more_2021_2023['특성별(2)'] = '4인 이상'
df_4_more_2021_2023.columns = ['2021', '2022', '2023', '특성별(2)']  # Ensure proper column alignment

# Merge both datasets into one row
df_4_more_combined = pd.merge(
    df_4_more_avg_2018_2020, df_4_more_2021_2023, on='특성별(2)', how='outer'
)

# Remove '4인', '5인 이상', and old '4인 이상' rows from the dataset
df_hmr_combined = df_hmr[~df_hmr['특성별(2)'].isin(['4인', '5인 이상', '4인 이상'])]
df_hmr_combined = pd.concat([df_hmr_combined, df_4_more_combined], ignore_index=True)

# Set up plot
plt.figure(figsize=(12, 8))
colors = ['b', 'r', 'g', 'c', 'm', 'y']
markers = ['o', 's', 'D', '^', 'v']
group_labels_combined = df_hmr_combined['특성별(2)'].unique()

# Plot each household size
for i, label in enumerate(group_labels_combined):
    spending_values = df_hmr_combined[df_hmr_combined['특성별(2)'] == label].iloc[:, 2:].values.flatten()
    
    # Per-person spending calculation
    if label == '4인 이상':
        spending_values = [float(value) / 4 if not pd.isna(value) else None for value in spending_values]
    elif label == '5인 이상':
        spending_values = [float(value) / 5 if not pd.isna(value) else None for value in spending_values]
    elif label == '3인':
        spending_values = [float(value) / 3 if not pd.isna(value) else None for value in spending_values]
    elif label == '2인':
        spending_values = [float(value) / 2 if not pd.isna(value) else None for value in spending_values]
    else:
        spending_values = [float(value) if not pd.isna(value) else None for value in spending_values]
    
    # Filter valid data points
    filtered_spending_values = []
    filtered_years = []
    for year, value in zip(years_hmr, spending_values):
        if value is not None:
            filtered_spending_values.append(value)
            filtered_years.append(year)

    # Plotting
    if filtered_spending_values:
        plt.plot(filtered_years, filtered_spending_values, marker=markers[i % len(markers)], linestyle='-', color=colors[i % len(colors)], label=label)

# Finalize plot
plt.title('연도별 가구원수별 1인당 월평균 간편식 구입 지출액 추이')
plt.xlabel('연도')
plt.ylabel('1인당 월평균 지출액 (원)')
plt.xticks(years_hmr, rotation=45)
plt.legend(title='가구원수별')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 2. 간편식 구입이유및 확인사항

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
file_path_reason = 'data/간편식_HMR__구입_이유_20241110181333.csv'
df_reason = pd.read_csv(file_path_reason)

# Extract data for visualization
reasons = df_reason.iloc[0, 1:].values
proportions = df_reason.iloc[1, 1:].replace(['-', '1인', '0.'], 0).astype(float).values
proportions = proportions[1:] 
reasons = [reason for reason in reasons if reason != '특성별(2)']

# Create a DataFrame for sorting
df_chart = pd.DataFrame({'구입 이유': reasons, '비중 (%)': proportions})

# Sort by '비중 (%)' in descending order
df_chart = df_chart.sort_values(by='비중 (%)', ascending=False)

# Horizontal bar chart
plt.figure(figsize=(10, 8))
plt.barh(df_chart['구입 이유'], df_chart['비중 (%)'], color='skyblue')  # Use barh for horizontal bars
plt.title('1인가구의 간편식 구입 이유 비중')
plt.ylabel('구입 이유')  # X축과 Y축이 반전되었으므로 여기가 Y축
plt.xlabel('비중 (%)')  # 여기가 X축
plt.xlim(0, max(df_chart['비중 (%)']) * 1.1)  # Set x-axis from 0 to slightly above the maximum proportion
plt.tight_layout()
plt.gca().invert_yaxis()  # Reverse the Y-axis to have the largest value on top
plt.show()
```

## 3. 자주 구입하는 품목

```python
file_path = 'data/주로_구입하는_간편식_HMR__품목_20241110184749.csv'
# Reload data with first two rows as headers for years and items
data =pd.read_csv(file_path, header=None)

years = data.iloc[0]
items = data.iloc[1]
data = data.drop([0, 1]).reset_index(drop=True)

# Rename columns based on years and items, using unique labels for easier handling
data.columns = [f"{year}_{item}" for year, item in zip(years, items)]

# Reshape the data to a long format, separating year and item from the combined column names
data_long = data.melt(var_name='연도_품목', value_name='비중')

# Split '연도_품목' into '연도' and '품목' columns for clarity
data_long[['연도', '품목']] = data_long['연도_품목'].str.split('_', expand=True)
data_long = data_long.drop(columns=['연도_품목']).dropna().reset_index(drop=True)

# Convert '연도' to numeric and '비중' to float, filling missing values with 0 for items not present in all years
data_long['연도'] = pd.to_numeric(data_long['연도'], errors='coerce')
data_long['비중'] = pd.to_numeric(data_long['비중'], errors='coerce').fillna(0)

# Pivot the data to prepare for line plotting with years as rows and items as columns
data_pivot = data_long.pivot_table(index='연도', columns='품목', values='비중', fill_value=0)

mean_data_pivot = data_pivot.mean()  # Calculate the mean 비중 for each item
filtered_columns = mean_data_pivot[mean_data_pivot >= 5].index  # Select items with mean 비중 >= 5%

# Filter the pivoted data to keep only these selected columns
filtered_data_pivot = data_pivot[filtered_columns]

# Plot the filtered data
plt.figure(figsize=(12, 8))
for column in filtered_data_pivot.columns:
    plt.plot(filtered_data_pivot.index, filtered_data_pivot[column], marker='o', label=column)

plt.xlabel('연도')
plt.ylabel('비중 (%)')
plt.title('5% 이상 품목의 연도별 비중 변화 (2018-2023)')
plt.legend(title='품목', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```
